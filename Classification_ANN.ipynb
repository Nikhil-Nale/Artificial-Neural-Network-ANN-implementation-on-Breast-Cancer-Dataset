```python
import numpy as npp
import pandas as pd
import pandas.util.testing as tm
#Data spliting
from sklearn.model_selection import train_test_split
#Keras api
from keras.models import Sequential
from keras.layers import Dense
from keras.utils.vis_utils import plot_model
```

```python
# Importing the data
from google.colab import files
file = files.upload()
```

```python
#Load the data into the pandas dataframe
df = pd.read_csv('data.csv')
df.head()
```

```python
df.shape
```

```python
#Data Cleaning
df.isna().sum()
```

```python
#Check the null columns
df[df["Unnamed: 32"].isnull()]
```

```python
#Drop the null column
df.dropna(axis=1, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
# Converting the categorical diagnosis column to numerical values because this is our target variable
df["diagnosis"] = npp.where(df["diagnosis"]=='M', 1, 0)
```

```python
df
```

```python
#Defining input variables X
X = df.iloc[:, 2:]
```

```python
X.shape
```

```python
#Defining the target variables Y
Y = df.iloc[:, 1:2]
```

```python
Y.shape
```

```python
#Splitting data into train and test dataset
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.30, random_state=30)
```

```python
#Defining ANN Sequencial model for tumar classification
classification_model = Sequential()
```

```python
classification_model.add(Dense(units=16, activation='relu', input_dim=30))
# Units indicated the output dimenssion of Dense layer
classification_model.add(Dense(units=8, activation='relu'))
classification_model.add(Dense(units=6, activation='relu'))
classification_model.add(Dense(units=1, activation='sigmoid'))
```

```python
classification_model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
print(classification_model.summary())
```

```python
classification_model.fit(x_train, y_train, batch_size=1, epochs=50)
# batch_size --> it is an integer number that indicated the number of samples per gradient update.
# epochs--> it is generally fine as one pass over the entire dataset.
```

```python
# lets quickly check our testing dataset
x_test
```

```python
# Now lets predict the target variable on test dataset Using our Sequential Model
prediction = classification_model.predict(x_test)
prediction = [1 if y>0.5 else 0 for y in prediction] #list
prediction
```

```python
# Converting prediction list to pandas dataFrame
pred = pd.DataFrame(prediction)
```

```python
# Now converting both pred dataframe and y_test dataframe to numpy ndarray to calculate prediction accuracy
```

```python
pred = pred.values
```

```python
y_test_ndarray = y_test.values
```

```python
# Prediction Accuracy
```

```python
total = 0
correct = 0 
wrong = 0

for i in range(len(pred)):
  total = total+1
  if (y_test_ndarray[i]==pred[i]):
    correct = correct+1
  else:
    wrong = wrong+1

print('Total = ' + str(total))
print('Correct= ' + str(correct) + ' - Percentage correct: ' + str(round((correct/total)*100, 2))+'%')
print('WRONG= ' + str(wrong) + 'Percantage wrong: ' + str(round((wrong/total)*100, 2)) + '%')
```
